```python
%sql
-- 1. Create Storage Credential
CREATE STORAGE CREDENTIAL `sc-adls-gani`
COMMENT 'Storage credential for ADLS Gen2 via Managed Identity';

-- 2. Create External Location for Bronze
CREATE EXTERNAL LOCATION `el-bronze-gani`
URL 'abfss://bronze@stdevelopmentgani.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `sc-adls-gani`)
COMMENT 'External location for Bronze layer';

-- 3. Create External Location for Silver
CREATE EXTERNAL LOCATION `el-silver-gani`
URL 'abfss://silver@stdevelopmentgani.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `sc-adls-gani`)
COMMENT 'External location for Silver layer';

-- 4. Create External Location for Gold
CREATE EXTERNAL LOCATION `el-gold-gani`
URL 'abfss://gold@stdevelopmentgani.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `sc-adls-gani`)
COMMENT 'External location for Gold layer';
```

```python
%sql
-- 1. Create Catalog
CREATE CATALOG IF NOT EXISTS dev
COMMENT 'Development catalog for F1 Analytics project';

-- 2. Create Schemas
CREATE SCHEMA IF NOT EXISTS dev.bronze
MANAGED LOCATION 'abfss://bronze@stdevelopmentgani.dfs.core.windows.net/'
COMMENT 'Bronze layer - raw ingested data';

CREATE SCHEMA IF NOT EXISTS dev.silver
MANAGED LOCATION 'abfss://silver@stdevelopmentgani.dfs.core.windows.net/'
COMMENT 'Silver layer - cleaned and transformed data';

CREATE SCHEMA IF NOT EXISTS dev.gold
MANAGED LOCATION 'abfss://gold@stdevelopmentgani.dfs.core.windows.net/'
COMMENT 'Gold layer - aggregated analytics ready data';
```
