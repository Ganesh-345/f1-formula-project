```python
from pyspark.sql import functions as f
file_name= ['constructors.json','constructors_flat']
# Due to the nested column in the json source files laoding from here
path_flat ="abfss://bronze@stdevelopmentgani.dfs.core.windows.net/formula1fullloadlanding/constructors/constructors_flat"
path = "abfss://bronze@stdevelopmentgani.dfs.core.windows.net/formula1fullloadlanding/constructors"

df_flat = spark.read.format('json')\
    .load(f"{path}/{file_name[0]}")

df_flat =df_flat.withColumn("ingestion_timestamp", f.current_timestamp())\
    .withColumn("source_file_name", f.col("_metadata.file_path"))

df_flat.write.format("parquet").mode('overwrite').save(path_flat)
```
