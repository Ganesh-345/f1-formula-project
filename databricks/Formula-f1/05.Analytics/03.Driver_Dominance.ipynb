```python
%sql

create or replace view vw_driver_dominance_view as
with driver_metric as (
    select
        driver_name,
        sum(race_starts) as race_starts,
        sum(total_wins) as total_wins,
        sum(total_podiums) as total_podiums,
        sum(case when standing_position = 1 then 1 else 0 end) as total_championships
    from vw_driver_standing_view
    group by driver_name
    having total_championships>1
)
select
    driver_name,
    race_starts,
    total_wins,
    total_podiums,
    total_championships,
    (total_championships * 100) + (total_wins * 10) + (total_podiums * 3) as greatness_score
from driver_metric
order by greatness_score desc



```
